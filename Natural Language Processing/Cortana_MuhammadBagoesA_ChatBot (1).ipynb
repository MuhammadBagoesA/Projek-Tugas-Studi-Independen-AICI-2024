```python
#1
import nltk
from nltk import ngrams
import random
from nltk.tokenize import word_tokenize

!gdown --id 15UqmiIm0xwh9mt0IYq2z3jHaauxQSTQT
nltk.download('punkt')

data = open('./irish-lyrics-eof.txt').read()
print(data)
```

```python
#tokenize kata
tokens = word_tokenize(data.lower())

#membuat trigram
trigrams = list(ngrams(tokens, 3))

#membuat dict
model = {}
```

```python
#mengisi model dengan trigram
for w1, w2, w3 in trigrams:
    if (w1, w2) not in model:
        model[(w1, w2)] = []
    model[(w1, w2)].append(w3)
```

```python
def generate_text(model, start, length=50):
    output = list(start)
    for _ in range(length):
        state = tuple(output[-2:])
        if state in model:
            next_word = random.choice(model[state])
            output.append(next_word)
        else:
            break
    return ' '.join(output)

start_tuple = tuple(tokens[:2])
```

```python
generated_text = generate_text(model, start_tuple, length=50)

print(generated_text)
```

```python
#2

import nltk
import random
import string
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
nltk.download('punkt')
nltk.download('wordnet')

# Membaca file dan mempersiapkan teks
f = open('Mod10_2.txt', 'r', errors='ignore')
raw = f.read()
raw = raw.lower()

# Tokenisasi kalimat dan kata
sent_tokens = nltk.sent_tokenize(raw)
word_tokens = nltk.word_tokenize(raw)

# Lemmatizer
lemmer = nltk.stem.WordNetLemmatizer()

def LemTokens(tokens):
    return [lemmer.lemmatize(token) for token in tokens]

remove_punct_dict = dict((ord(punct), None) for punct in string.punctuation)

def LemNormalize(text):
    return LemTokens(nltk.word_tokenize(text.lower().translate(remove_punct_dict)))

# Greeting response setup
GREETING_INPUTS = ("hello", "hi", "greetings", "sup", "what's up", "hey")
GREETING_RESPONSES = ["hi", "hey", "*nods*", "hi there", "hello", "I am glad! You are talking to me"]

def greeting(sentence):
    for word in sentence.split():
        if word.lower() in GREETING_INPUTS:
            return random.choice(GREETING_RESPONSES)

# Fungsi untuk memberikan response chatbot
def response(user_response):
    robo_response = ''
    sent_tokens.append(user_response)  # Menambahkan respons pengguna ke dalam list kalimat
    TfidfVec = TfidfVectorizer()  # Inisialisasi TfidfVectorizer
    tfidf = TfidfVec.fit_transform(sent_tokens)  # Menghitung matriks TF-IDF
    vals = cosine_similarity(tfidf[-1], tfidf)  # Menghitung kesamaan cosine antara kalimat terakhir dan kalimat lainnya
    idx = vals.argsort()[0][-2]  # Mendapatkan indeks dari kalimat yang paling mirip
    flat = vals.flatten()
    flat.sort()
    req_tfidf = flat[-2]

    if(req_tfidf == 0):
        robo_response = "I am sorry! I don't understand you"
    else:
        robo_response = sent_tokens[idx]

    sent_tokens.remove(user_response)  # Menghapus respons pengguna setelah diproses
    return robo_response

# Program utama untuk interaksi
if __name__ == "__main__":
    flag = True
    print("Cortana: My name is Cortana. I will answer your queries about Chatbots. If you want to exit, type Bye!")

    while flag:
        user_response = input("You: ")  # Menunggu input dari pengguna
        user_response = user_response.lower()

        if user_response != 'bye':
            if user_response == 'thanks' or user_response == 'thank you':
                flag = False
                print("Cortana: You are welcome..")
            else:
                if greeting(user_response) is not None:
                    print("Cortana: " + greeting(user_response))
                else:
                    print("Cortana: ", end="")
                    print(response(user_response))
        else:
            flag = False
            print("Cortana: Bye! Take care..")
```
