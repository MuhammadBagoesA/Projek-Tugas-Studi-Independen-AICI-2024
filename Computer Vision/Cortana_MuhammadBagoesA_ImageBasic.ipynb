```python
#1
import cv2
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt

# Path ke gambar
image_path = 'tesbutawarna.jpg'

# Membaca gambar menggunakan OpenCV
image = cv2.imread(image_path)

# Memeriksa apakah gambar berhasil dimuat
if image is None:
    print(f"Error: Gambar tidak ditemukan atau gagal dibaca di path: {image_path}")
else:
    # Mengonversi gambar ke ruang warna HSV
    hsv_image = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)

    # Memilih channel Hue dari HSV
    hue_channel = hsv_image[:, :, 0]

    # Mengonversi channel menjadi gambar PIL
    pil_image = Image.fromarray(hue_channel)

    # Menampilkan gambar dengan matplotlib
    plt.imshow(pil_image, cmap='gray')
    plt.axis('off')
    plt.show()
```

```python
#2
import numpy as np
img = cv2.imread("doggo.jpg")

img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

#tulis I
font = cv2.FONT_HERSHEY_SIMPLEX
cv2.putText(img,text='I', org=(30,30), fontFace=font, fontScale=1,color=(255,255,255),thickness=2,lineType=cv2.LINE_AA)

#gambar <3
cv2.circle(img, center=(30,60), radius=10, color=(255,0,0), thickness=-1)
cv2.circle(img, center=(45,60), radius=10, color=(255,0,0), thickness=-1)
vertices = np.array([[20,60],[55,60],[38, 80]],np.int32)
pts = vertices.reshape((-1,1,2))
cv2.polylines(img,[pts],isClosed=True, color=(255,0,0),thickness=2)
cv2.fillPoly(img, [pts], color=(255, 0, 0))

#tulis Dog
font = cv2.FONT_HERSHEY_SIMPLEX
cv2.putText(img,text='DOG', org=(10,120), fontFace=font, fontScale=1,color=(255,255,255),thickness=2,lineType=cv2.LINE_AA)
plt.imshow(img)
```
