```python
#1
import sklearn.datasets
import pandas as pd

data = sklearn.datasets.load_wine()

data_wine = pd.DataFrame(data.data, columns=data.feature_names)
data_wine["target"] = data.target

data_wine
```

```python
from sklearn.model_selection import train_test_split

X = data_wine.drop("target", axis=1)
y = data_wine["target"]

X_train, X_val_and_test, y_train, y_val_and_test = train_test_split(X, y, test_size=0.3)
X_test, X_val, y_test, y_val = train_test_split(X_val_and_test, y_val_and_test, test_size=0.5)
```

```python
from keras.models import Sequential
from keras.layers import Dense

model = Sequential([
    Dense(10, activation='relu', input_shape=(13,)),
    Dense(10, activation='relu'),
    Dense(1, activation='sigmoid'),
])
```

```python
model.compile(
    optimizer='sgd',
    loss='binary_crossentropy',
    metrics=['accuracy']
)
```

```python
model.fit(
    X_train, y_train,
    batch_size=32,
    epochs=100,
    validation_data=(X_val, y_val)
)
```

```python
accuracy = model.evaluate(X_test, y_test)[1]
print(f"Accuracy: {accuracy}")
```
